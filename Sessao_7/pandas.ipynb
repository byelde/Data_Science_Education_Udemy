```python
import pandas as pd
import numpy as np
```

```python
# Carregar arquivo como dataframe
dados = pd.read_csv('dados/Credit.csv')

# Formato
dados.shape
```

```python
# Resumo de estatisticas numéricas
dados.describe()
```

```python
# Primeiros registros
display(dados.head(3))

# Ultimos registros
display(dados.tail(3))
```

```python
# Filtrar por nome da coluna
display(dados['duration'])  """UM""" # par de chaves
display(dados[['duration']]) """DOIS""" # pares de chaves
```

```python
# Filtrar linha por indice
display(dados.loc[1:3]) # Funciona

# Filtrar linhas específicas
display(dados.loc[[1,3,7]]) # Funciona

# dados.loc(parametro)
# Parametro: uma lista (mais um par de chaves) para linhas especóificos específicas
# ou um intervalo, sem chaves, com : para um intervalo de linhas

'''
display(dados.loc[1:3]) >>NÃO FUNCIONA<<
display(dados.loc[[1:3]) >>NÃO FUNCIONA<<
'''
```

```python
# Filtro

display(dados.loc[dados['credit_history'] == "'existing paid'"].head())
# display(dados.loc[dados['credit_amount'] > 3271.258].head()) # Escondido
```

```python
# Criar outros dataframes com filtro
credit_amount = dados.loc[dados['credit_amount'] > 3271.258]
display(credit_amount.head())
```

```python
# Filtro em colunas selecionadas

credito = dados[['age','credit_history', 'credit_amount']].loc[dados['age'] <= 24]
# Dados de histórico de crédito e quantidade de crédito de pessoas com 24 anos ou menos
display(credito)
```

```python
# Series

# Com uma lista
print('LISTA')
s1 = pd.Series([11,12,13,14,15])
print(s1)

# Com array do numpy
print('\n\nNUMPY')
a = np.arange(5,10)
s2 = pd.Series(a)
print(s2)

# Com dados de colunas
print('\n\nCOLUNA')
b = dados['duration']
s3 = pd.Series(b)
print(s3)
```

```python
# Renomear e manter alterações sem precisar usar 'obj = obj.alteração'

dados.rename(columns={'duration':'Duracao', 'credit_history':'HistCred'}, inplace=True) # IMPLACE MANTEM ALTERAÇÔES
display(dados.head(3))
```

```python
# Excluir coluna
dados.drop('other_payment_plans', axis=1, inplace=True)
display(dados)
```

```python
# Verificar dados nulos

display(dados.isnull().head(3))
display(dados.isnull().sum())
```

```python
# Retirar dados NaN
display(dados.dropna())

# Substituir dados NaN
dados.fillna(0, inplace=True)
```

```python
# iLoc

display(dados.iloc[0:3,0:4]) # mostra linha de 0 a 3 e colunas 0 a 4
display(dados.iloc[[1,3,5], 0:4]) # mostra linhas 1,3,5 e colunas de 0 a 4
display(dados.iloc[[1,3,5],[9,2,6]]) # mostra linhas 1,3,5 e colunas 9,2 e 6 
```
